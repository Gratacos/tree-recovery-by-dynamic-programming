```python
import networkx as nx

ids = './datasets/truncated_datasets/' + dataset + '/ids/'
small = './datasets/truncated_datasets/' + dataset + '/'
large = './datasets/full_datasets/' + dataset + '/'
gt = np.load('D:\\Documents\\Reps\\Erosion\\wide\\ground_truth.npy', allow_pickle=True)[()]
for gname in gt: 
    gt[gname] = nx.relabel_nodes(gt[gname][1], {n:n-1 for n in sorted(gt[gname][1])})
    gt[gname].graph['root'] = gt[gname].graph['root'][0] - 1
cfile_full = {gname:[] for gname in cfile}
for gname in cfile:
    g = skh.ply_to_graph(os.path.join(small, gname + '.ply'))
    # G = skh.ply_to_graph(os.path.join(large, gname + '.ply'))
    G = gt[gname]
    inds = []
    with open(os.path.join(ids, gname + '.txt')) as f:
        lines = f.readlines()
        for line in [l for l in lines if l != '']:
            inds.append(int(line))
            
    junctions = [e for e in g.edges() if nx.degree(g, e[0]) > 2 or nx.degree(g, e[1]) > 2]
    
    cfile_full[gname] = [0]*len(G)
    for i in range(len(cfile[gname])):
        ogi = inds[i]
        #print(len(cfile_full[gname]), ogi)
        cfile_full[gname][ogi] = cfile[gname][i]
        # print( ([G.edges[e]['name'] for e in G.edges(ogi, keys=True)]), (cfile[gname][i][0][0]))
        #assert (len([G.edges[e]['name'] for e in G.edges(ogi, keys=True)]) == len(cfile[gname][i][0][0])), str([G.edges[e]['name'] for e in G.edges(ogi)]) + '!=' + str(cfile[gname][i][0][0])
    for i in range(len(cfile_full[gname])):
        if cfile_full[gname][i] == 0:
            cfile_full[gname][i] = gen_full(sorted([G.edges[e]['name'] for e in G.edges(i, keys=True)]))
```

```python
# Save full dset:
for gname in gt:
    for n in gt[gname]:
        if n != gt[gname].graph['root']:
            gt[gname].nodes[n]['root'] = 0
        else:
            gt[gname].nodes[n]['root'] = 1 
        
        keys = list(gt[gname].nodes[n].keys())
        for k in keys:
            if k not in {'x', 'y', 'z', 'root'}:
                del gt[gname].nodes[n][k]
    
    skh.graph_to_ply(nx.Graph(gt[gname]), os.path.join(large, gname+'_l.ply'))
    
    # Save costs:
    cf_txt = ''
    for cf in cfile_full[gname]:
        for l in cf:
            for attr in l[:-1]:
                for elem in attr:
                    cf_txt += str(elem)
                    cf_txt += ' '
                cf_txt = cf_txt[:-1] + ','
            #print(l)
            cf_txt += str(l[3]) + ';'
        cf_txt += '\n'
    
    print('Saving costs to ', os.path.join(large, gname+'_costs.txt'))
    with open(os.path.join(large, gname+'_costs.txt'), 'x') as f:
        f.write(cf_txt)
```
