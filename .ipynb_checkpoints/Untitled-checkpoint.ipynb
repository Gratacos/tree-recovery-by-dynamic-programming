```python
import numpy as np
import os
import sys
```

```python
fold = './rice/'
fnames = [f for f in os.listdir(fold) if f.endswith('costs.txt')]
```

```python
new_file_ext = 'costs_no_deg2.txt'

cfile = {}
for fname in fnames:
    path = os.path.join(fold, fname)
    csts = []
    with open(path, 'r') as f:
        lines = f.readlines()
        for line in lines:
            line = line.replace('\n', '')
            tokens = [c for c in line.split(';') if c != '']
            csts.append([])
            for tok in tokens:
                elms = [t for t in tok.split(',') if t != '']
                assign = [int(e) for e in elms[0].split(' ') if e != '']
                partition = [int(e) for e in elms[1].split(' ') if e != '']
                directions = [int(e) for e in elms[2].split(' ') if e != '']
                labels = [int(e) for e in elms[3].split(' ') if e != '']
                cost = float(elms[4])
                csts[-1].append([assign, partition, directions, labels, cost])
    cfile[fname.split('.')[0][:8]] = csts
```

```python
csts = cfile['graph_01_costs']
for cands in csts:
    for cand in cands:
        if len(cand[0]) == 2:
            print(cand)
```
